```python
import pandas as pd
import matplotlib.pyplot as plt

veriSeti = pd.read_csv('data.csv')


print(veriSeti.columns)


yayinYillari = veriSeti['releaseYear']
turler = veriSeti['genres']

# Her yıl için drama türündeki içerik hesabı yaptım.
drama_sayisi = turler.str.contains('Drama').groupby(yayinYillari).sum()

plt.figure(figsize=(20, 7))
plt.bar(drama_sayisi.index, drama_sayisi.values)
plt.xlabel('Yayın Yılları')
plt.ylabel('Drama Türündeki İçerik Sayısı')
plt.title('Yıllara Göre Drama Türündeki İçerik Sayısı')
plt.xticks(rotation=90)
plt.show()
```
